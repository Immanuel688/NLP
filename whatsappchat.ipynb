```python
# Now create a function for sentiment analyser and the input will be the column of the pre-processes dataframe
```

```python
def Sentiment(dataframe,columnName):
    import nltk
    #nltk.download('vader_lexicon')
    from nltk.sentiment.vader import SentimentIntensityAnalyzer
    nltk.download('vader_lexicon')
    sid= SentimentIntensityAnalyzer()
    dataframe.dropna(inplace=True)
    dataframe['scores'] = dataframe[columnName].apply(lambda commentText: sid.polarity_scores(commentText))
    dataframe['compound']= dataframe['scores'].apply( lambda score_dict:score_dict['compound'])
    dataframe['neutral']=dataframe['scores'].apply(lambda score : score['neu'])
    dataframe['positive']=dataframe['scores'].apply(lambda score : score['pos'])
    dataframe['comp_score']=dataframe['compound'].apply(lambda c : 'pos' if c>=0 else 'neg')
    comp_table=pd.DataFrame(dataframe['comp_score'].value_counts()).reset_index()
    return dataframe, comp_table
```
