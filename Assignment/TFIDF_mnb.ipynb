```python
import pandas as pd

def read_flirt_dataset(flirt_dataset):
    rows = []
    with open(flirt_dataset, encoding="utf8") as f:
        header = f.readline().strip().split(",")  # read first line as header
        for line in f:
            parts = line.strip().split(",")
            # first three are fixed: date, time, speaker
            date = parts[0]
            time = parts[1]
            speaker = parts[2]
            flirt = parts[-1]   # last column
            # everything in between is the message
            message = ",".join(parts[3:-1])
            rows.append([date, time, speaker, message, flirt])
    
    return pd.DataFrame(rows, columns=header)

# Usage
df = read_flirt_dataset("flirt_dataset.txt")
print(df)
```

```python
x=df['message']
y=df['flirt']
```

```python
#TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
# NOW WE HAVE SEPERATED TRAIN AND TEST NOW WE HAVE TO CONVERT THE TEXT TO VECTORS
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, HashingVectorizer # Helps converting text to vectors(numbers)
```

```python
tf_idf=TfidfVectorizer(stop_words='english')
tf_idf
```

```python
tf_idf_train=tf_idf.fit_transform(X_train)
print(tf_idf_train)
```

```python
tf_idf_test=tf_idf.transform(X_test)
print(tf_idf_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf=MultinomialNB()
clf.fit(tf_idf_train,y_train)
```

```python
pred=clf.predict(tf_idf_test)
pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_test,pred)
score
```

```python
print(f"accuracy:   {score:.3f}")
```

```python
report=classification_report(y_test, pred)
print(report)
```
