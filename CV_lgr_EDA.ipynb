```python
import pandas as pd

def read_flirt_dataset(flirt_dataset):
    rows = []
    with open(flirt_dataset, encoding="utf8") as f:
        header = f.readline().strip().split(",")  # read first line as header
        for line in f:
            parts = line.strip().split(",")
            # first three are fixed: date, time, speaker
            date = parts[0]
            time = parts[1]
            speaker = parts[2]
            flirt = parts[-1]   # last column
            # everything in between is the message
            message = ",".join(parts[3:-1])
            rows.append([date, time, speaker, message, flirt])
    
    return pd.DataFrame(rows, columns=header)

# Usage
df = read_flirt_dataset("flirt_dataset.txt")
print(df)
```
